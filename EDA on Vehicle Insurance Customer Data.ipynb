### 3. Replace all null values for numeric columns by mean. and Replace all null values for Categorical value by mode. 

```python
cd=pd.read_csv(r'D:/python/customer_details.csv', header=0, names=['customer_id','Gender','age', 'driving_licence_present','region_code','previously_insured','vehicle_age','vehicle_damage'], index_col='customer_id')
cp=pd.read_csv("D:\python\customer_policy_details.csv", header=0,index_col='customer_id', names=['customer_id','annual premium','sales channel code','vintage','response'])
```

```python
# filling the mean of each integer column at the place of null values
cd['age']=cd['age'].fillna(cd['age'].mean())
cd['previously_insured']=cd['previously_insured'].fillna(cd['previously_insured'].mean())
cd['region_code']=cd['region_code'].fillna(cd['region_code'].mean())
cd['driving_licence_present']=cd['driving_licence_present'].fillna(cd['driving_licence_present'].mean())


# converting the object type in string as they have mix values so not allowing to fill it with mode 
cd['Gender']=cd['Gender'].astype(str)
cd['vehicle_age']=str(cd['vehicle_age'])
cd['vehicle_damage']=str(cd['vehicle_damage'])
# now filling it with mode.
cd['Gender']=cd['Gender'].fillna(cd['Gender'].mode())
cd['vehicle_damage']=cd['vehicle_damage'].fillna(cd['vehicle_damage'].mode())
cd['vehicle_age']=cd['vehicle_age'].fillna(cd['vehicle_age'].mode())
cd['age'].isnull().sum()
cd.info()
```

```python
# filled all the values of the DF with the mean 
cp.info()
cp['annual premium']=cp['annual premium'].fillna(cp['annual premium'].mean())

cp['sales channel code'].fillna(cp['sales channel code'].mean(), inplace=True)
cp['vintage'].fillna(cp['vintage'].mean(),inplace=True)
cp['response'].fillna(cp['response'].mean(),inplace=True)
cp.info()
```

# ii. Outliers

Generate a summary of count of all the outliers column wise
Replace all outlier values for numeric columns by mean. 
(Hint1: for outlier treatment use IQR method as follows:

For example: for a column X calculate Q1 = 25th percentile and Q3 = 75th percentile then IQR = Q3 – Q1 ) then to check outlier, anything lower than a Q1 – 1.5IQR or greater than Q3 + 1.5 IQR would be an outlier

Hint2: For getting percentile value, explore pd.describe() function)

iii. White spaces

Remove white spaces
iv. case correction(lower or upper, any one) 

v. Convert nominal data (categorical) into dummies 

for future modeling use if required
vi. Drop Duplicates (duplicated rows)

```python
q1=cd['age'].quantile(0.25)
q3=cd['age'].quantile(0.75)
IQR=q3-q1
age_min_outliner=(q1-1.5*IQR)
age_max_outliner=(q3+1.5*IQR)

print("Minimum age",age_min_outliner,"Maximum age",age_max_outliner)

cd.loc[cd['age']>=age_max_outliner, 'age']
cd.loc[cd['age']>=age_max_outliner, 'age'].sum()
cd.loc[cd['age']>=age_max_outliner, 'age']=cd['age'].mean()
```

```python
cd.loc[cd['age']>=85,'age']
```

```python
# iii. White spaces  Remove white spaces

cd['Gender'] = cd['Gender'].str.strip()
cd['vehicle_age'] = cd['vehicle_age'].str.strip()
cd['vehicle_age'] = cd['vehicle_age'].str.replace(' ', '', regex=True)
cd
print(cd['vehicle_age'])
```

```python
# iv. case correction(lower) 
cd['Gender']=cd['Gender'].str.lower()
cd
# iv. case correction(Upper)
cd['Gender']=cd['Gender'].str.upper()
cd
# iv. case correction(capitalize)
cd['Gender']=cd['Gender'].str.capitalize()
cd
```

```python
#v. Convert nominal data (categorical) into dummies 
#for future modeling use if required
print(cd['region_code'].count())
dummies=pd.get_dummies(cd,columns='region_code', drop_first=True)
print(dummies)
```

```python
#vi. Drop Duplicates (duplicated rows)
cd.dropna(how='any')
cd.info()
cd
```
